```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = fetch_openml('car', as_frame=True)
```

```python
data 
#feature encoding is done when datasets contains categorical data and are not numerical 
```

```python
columns_to_encode=['lug_boot','safety']
encoder=OrdinalEncoder(
  categories=[
    ['small','med','big'],
    ['low','med','high'],
  ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```
